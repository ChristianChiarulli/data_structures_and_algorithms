# Asymptotic Analysis with: Big-O, Big-$ \Omega $, Big-$ \Theta $

## Comparing Growth Rates

These are the seven fundemental functions used in algorithm analysis. ($a > 1$)

|Constant|Logarithm|Linear|Log Linear|Quadratic|Cubic|Exponential|
|--------|---------|------|----------|---------|-----|-----------|
|$1$     |$log (n)$|$n$   |$nlog(n)$ |$n^2$    |$n^3$|$a^n$      |

Data structure operations ideally should run in times proportional to the *constant* or *logarithm* function.

Algorithms should run in linear or *n-log-n* time.

Algorithms with *quadratic* and *cubic* run times are much less practical.

Algorithms with *exponential* run times are mostly infeasible for all but the smallest input sizes.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(1, 13, .3)

growth_dict = {
    'Constant': np.ones(n.shape),
    'Logarithmic': np.log(n),
    'Linear': n,
    'Log Linear': n*np.log(n),
    'Quadratic': n**2,
    'Cubic': n**3,
    'Exponential': 2**n    # Note a is 2
}

plt.figure(figsize=(20,10))

plt.ylim(0, 30)  # Comment out to show how bad exponential can be

labels = list(growth_dict.keys())
big_o = list(growth_dict.values())
    
for i in range(len(labels)):
    plt.plot(n, big_o[i],label = labels[i])
    
plt.legend()
plt.ylabel('Running Time')
plt.xlabel('Input Size n')
plt.show()
```

## Big-O

Let $f(n)$ and $g(n)$ be functions mapping positive integers to positive real numbers.

We say $f(n)$ is $O(g(n))$ if there is a real constant $c > 0$ and an integer constant $n_0 \geq 1$ such that:

$f(n) \leq cg(n)$, for $n \geq n_0$

This is often referred to as Big-O notation and is sometimes pronounced as $f(n)$ is Big-O of $g(n)$

Remember Big-O is an ***UPPER BOUND***

```python
n = np.arange(1, 10, .3)

big_o_dict= {
    'f(n)': 10*n**2 + 10*n + 300,
    'cg(n)': 20*n**2,
    
    #'ch(n)': 2*n**3 + 10*n + 300,    # Overshooting
    #'ck(n)': 500*np.log(n),
}

plt.figure(figsize=(20,10))
plt.ylim(0, 1750)

labels = list(big_o_dict.keys())
big_o = list(big_o_dict.values())
    
for i in range(len(labels)):
    plt.plot(n, big_o[i],label = labels[i])
    
plt.legend()
plt.ylabel('Running Time')
plt.xlabel('Input Size n')
plt.show()
```

The above example illustrates the "big-Oh" notation. The function $f(n)$ is $O(g(n))$, since $f(n) \leq c \cdot g(n)$ when $n \geq n_0$

As long as we can find a $c$ that overcomes the original function then all of the other terms clearly can be dropped as well as the leading coefficient

Remember we care what happens as we grow towards infinity, so unless we are dealing with a constant factor like $10^{100}$ then we're fine

## Characterize Functions in Simplest Terms Possible

It is good practice to use big-Oh notation to characterize a function as closely as possible. 

You could say that the function $f(n) = 5n^3 + 7n$ is $O(n^6)$, $O(n^4)$ or even $O(n^{100})$. 

But we can choose an arbritrary $c$ of $20$ and $n^3$ this will ensure we grow faster, since: $5n^3 + 7n \leq 20n^3$, for $n \geq n_0$

So it is more accurate to say $f(n)$ is $O(n^3)$. (But nothing is forcing you to do this this is where big-$\Theta$ comes in)

Why do we drop the constant c? We drop the constant because it was completely arbritrary, We could have used $c = 21, 22, 23, 100,$ etc.. 

So it does not make sense to include the constant because we care about modeling behavior.

## Big-$ \Omega $

Let $f(n)$ and $g(n)$ be functions mapping positive integers to positive real numbers.

We say $f(n)$ is $\Omega(g(n))$ if there is a real constant $c > 0$ and an integer constant $n_0 \geq 1$ such that:

$f(n) \geq cg(n)$, for $n \geq n_0$

Remember Big-$\Omega$ is a ***LOWER BOUND***

```python
n = np.arange(1, 20, .3)  ## Change 20 to a much higher value to show that g(n)

big_omega_dict= {
    'f(n)': n**2 + 10*n,
    'cg(n)': .5*n**2
}

plt.figure(figsize=(20,10))

labels = list(big_omega_dict.keys())
big_o = list(big_omega_dict.values())
    
for i in range(len(labels)):
    plt.plot(n, big_o[i],label = labels[i])
    
plt.legend()
plt.ylabel('Running Time')
plt.xlabel('Input Size n')
plt.show()
```

The above example illustrates the big-$\Omega$ notation. The function $f(n)$ is $\Omega(g(n))$, since $f(n) \geq c \cdot g(n)$ when $n \geq n_0$

As long as we can find a $c$ that prevents $g(n)$ from overcoming $f(n)$ then we have found a $g(n)$ capable of modeling the correct behavior.

Remember we care what happens as we grow towards infinity.

Also by the way we could just choose $g(n)=1$, but again we would like to be more accurate.

## Big-$ \Theta $

Let $f(n)$ and $g(n)$ be functions mapping positive integers to positive real numbers.

We say $f(n)$ is $\Theta(g(n))$ if there is a real constant $c > 0$ and an integer constant $n_0 \geq 1$ such that:

$f(n) \geq cg(n)$, for $n \geq n_0$

Remember Big-$\Theta$ refers to a ***UPPER BOUND*** & ***LOWER BOUND***

```python
n = np.arange(1, 10, .3) # Increase 10 to 100, 1000, 10,000 notice it is always bounded

big_o_dict= {
    'f(n)': 10*n**2 + 10*n + 300,
    'cg(n)_1': 20*n**2,
    'cg(n)_2': 5*n**2,
}

plt.figure(figsize=(20,10))

plt.ylim(0, 1750)        # Comment out to let functions grow past 1750

labels = list(big_o_dict.keys())
big_o = list(big_o_dict.values())
    
for i in range(len(labels)):
    plt.plot(n, big_o[i],label = labels[i])
    
plt.legend()
plt.ylabel('Running Time')
plt.xlabel('Input Size n')
plt.show()
```
