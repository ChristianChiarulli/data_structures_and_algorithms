```python
import matplotlib.pyplot as plt
import numpy as np
import math

n = np.arange(1, 10**3, 0.05)

fcns_dict= {
    '8n': 8*n,
    '4nlogn': 4*n*np.log2(n),
    '$2n^2$': 2*n**2,
    '$n^3$': n**3,
#   '$2^n$': 2**n # Computationally intensive
}

plt.figure(figsize=(20,10))

labels = list(fcns_dict.keys())
fcns = list(fcns_dict.values())
    
for i in range(len(labels)):
    plt.plot(n, fcns[i], label = labels[i])
    
plt.legend()
plt.ylabel('Running Time')
plt.xlabel('Input Size n')
plt.yscale('log')
plt.xscale('log')
plt.xlim((1, 10**3))
plt.show()
```

#### R-3.2

To determine $n_0$ such that A is better than B for n $\geq$ $n_0$, we can determine the point where A and B intersect and $8nlogn \leq 2n^2$.

```python
n = np.arange(15, 17, 0.05)

fcns_dict= {
    'A': 8*n*np.log2(n),
    'B': 2*n**2,
}

plt.figure(figsize=(20,10))

labels = list(fcns_dict.keys())
fcns = list(fcns_dict.values())
    
for i in range(len(labels)):
    plt.plot(n, fcns[i], label = labels[i])
    
plt.legend()
plt.ylabel('Running Time')
plt.xlabel('Input Size n')
plt.xlim((15, 17))
plt.show()
```

$n_0 = 16$ from the plot since $A$ is better than $B$ for $n \geq 16$.

#### R-3.3

To determine $n_0$ such that A is better than B for n $\geq$ $n_0$, we can determine the point where A and B intersect and $40n^2 \leq 2n^3$.

```python
n = np.arange(18, 25, 1)

fcns_dict= {
    'A': 40*n**2,
    'B': 2*n**3,
}

plt.figure(figsize=(20,10))

labels = list(fcns_dict.keys())
fcns = list(fcns_dict.values())
    
for i in range(len(labels)):
    plt.plot(n, fcns[i], label = labels[i])
    
plt.legend()
plt.ylabel('Running Time')
plt.xlabel('Input Size n')
plt.xlim((18, 25))
plt.show()
```

$n_0 = 20$ from the plot since $A$ is better than $B$ for $n \geq 20$.

We can also determine the point of intersection by setting the expressions equal to each other, i.e.,

$$ 40n_0^2 = 2n_0^3 $$
$$ 40 = 2n_0 $$
$$ n_0 = 20 $$

#### R-3.5

The plot of the function $n^c$ is a straight line with a slope c on a log-log scale because if $$f(n) = n^c$$ and we take the log of both sides we get $$log(f(n)) = log(n^c)$$ $$log(f(n)) = clog(n)$$ which is in the form of a straight line, i.e., y = mx. 

Here, $y = log(f(n))$, $m = c$, and $x = log(n)$. 

Below is an example plot with c = 5.

```python
n = np.arange(1, 10**3, 1)

c = 5

fcn = n**c

plt.figure(figsize=(20,10))
    
plt.plot(n, fcn)
    
plt.title("Log-Log Scale")
plt.ylabel('Running Time')
plt.xlabel('Input Size n')
plt.yscale('log')
plt.xscale('log')
plt.xlim((1, 10**3))
plt.show()
```

#### R-3.10

If $d(n)$ is $O(f(n))$, then there is a real constant $c_1 > 0$ and an integer constant $a_0 \geq 1$ such that 

$ d(n) \leq c_1f(n)$, for $n \geq a_0$

Similarly, if $e(n)$ is $O(g(n))$, then there is a real constant $c_2 > 0$ and an integer constant $b_0 \geq 1$ such that

$ e(n) \leq c_2g(n)$, for $n \geq b_0 $

So, $d(n)e(n) \leq c_1f(n)c_2g(n)$, for $n \geq max(a_0, b_0)$

$d(n)e(n) \leq cf(n)g(n)$, for $n \geq max(a_0, b_0)$, where $c = c_1c_2$

Therefore, $d(n)e(n)$ is $O(f(n)g(n))$

#### R-3.18

$2^{n+1} = 2*2^n$

$\implies$ there is some constant $c > 0$, for $n \geq n_0$ such that

$2*2^n \leq c2^n$ 

E.g., let $c = 3$ and $n_0 = 1$

$2*2^1 \leq 3*2^1$

$4 \leq 6$

which means $2^{n+1}$ is $O(2^n)$

#### R-3.20

Let $f(n)$ and $g(n)$ be functions mapping positive integers to positive real numbers.

If there's a real constant $c > 0$ and an integer constant $n_0 \geq 1$ such that 

$f(n) \geq cg(n)$, for $n \geq n_0$ 

then $f(n)$ is $\Omega(g(n))$

Let $f(n) = n^2$, $g(n) = nlogn$, $c = 1$, and $n_0 = 2$

$2^2 \geq (1)(2)log(2)$

$4 \geq (2)(1)$

$4 \geq 2 $

$\therefore n^2$ is $\Omega(nlogn)$ 

#### R-3.23

```python
def example1(S):
    n = len(S)
    total = 0
    for j in range(n): # loop from 0 to n-1
        total += S
    return total
```

* Initializing the variables $n$ and $total$ takes $O(1)$ time each.

* The for loop is controlled by the counter $j$ and the maintenance of the counter is done by the range iterator which executes $n$ times, so the loop takes a total of $O(n)$ time.

* The body of the loop executes $n$ times, so the statement $total += 5$ is executed $n$ times with each execution of the statement taking $O(1)$ time, so the overall contribution is $O(n)$ time.

* The return statement takes $O(1)$ time.

The running time of $example1$ is given by the sum of $1 + 1 + n + n + 1 = 2n + 1$ which by Proposition 3.9 is $O(n)$.
